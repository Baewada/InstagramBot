```python
from instapy import InstaPy 

session = InstaPy(username="<username>", password="<password>", headless_browser=True)#set your username and passwords and set 'headless_browser = false' For GUI experience on browser  
session.login()
session.set_quota_supervisor(enabled=True, peak_comments_daily=100, peak_comments_hourly=10)
session.set_use_clarifai(enabled=True, api_key='<your_api_key>')
session.clarifai_check_img_for(['nsfw'])
session.set_relationship_bounds(enabled= True, min_followers= 800, max_followers= 1000)
session.like_by_tags(['cybersecurity', 'programming'], amount=5)#target Hashtags
session.set_dont_like(["naked", "nsfw"])#hashtags not to like 
session.set_do_follow(True, percentage=50)
session.set_do_comment(True, percentage=50)
session.set_comments(["Nice!", "Thanks4sharing"])#comments on posts
session.end()
```
